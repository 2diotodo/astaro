# TEXT파일의 데이터를 CSV파일로 변형.

```python
# Mount Google Drive to access the data file
from google.colab import drive
drive.mount('/content/drive')

import csv

# 파일 경로 지정
file_path = '/content/drive/MyDrive/astaro/data.txt'

# CSV 파일로 저장할 파일 경로 지정
output_path = '/content/drive/MyDrive/astaro/data.csv'

# CSV 파일에 저장할 헤더 지정
header = ["text", "result"]

# CSV 파일에 저장할 데이터를 담을 리스트
data = []

# TEXT 파일 읽어오기
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        # 끝의 두 글자가 '|0'이면 0, 그렇지 않으면 1로 결과값 설정
        if line.endswith("|0\n"):
            result = "0"
            text = line.strip()[:-2]  # 끝의 '|0\n'을 제거한 텍스트
        else:
            result = "1"
            text = line.strip()[:-2]  # 끝의 '|1\n'을 제거한 텍스트

        # CSV 파일에 저장할 데이터 리스트에 추가
        data.append([text, result])

# CSV 파일 쓰기
with open(output_path, "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)

print("CSV 파일로 변환 완료")
```
